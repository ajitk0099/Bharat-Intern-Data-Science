```python
# Changing Value for "Male, Female" string values to numeric values , male=1 and female=2
def getNumber(str):
    if str=="male":
        return 1
    else:
        return 2
    
titanic["Gender"]=titanic["Sex"].apply(getNumber)

titanic.head()
```

```python
del titanic["Sex"]
titanic.head()
```

```python
titanic.isnull().sum()
```

```python
meanS= titanic[titanic.SibSp==1].Age.mean()
meanS
```

```python
titanic["age"]=np.where(pd.isnull(titanic.Age) & titanic["SibSp"]==1  ,meanS, titanic["Age"])
titanic.head()
```

```python
titanic.isnull().sum()
```

```python
meanNS=titanic[titanic.SibSp==0].Age.mean()
```

```python
meanNS
```

```python
titanic.age.fillna(meanNS,inplace=True)
```

```python
titanic.head()
```

```python
titanic.isnull().sum()
```

```python
del titanic["Age"]
titanic.head()
```

```python
# Finding the number of people who have survived 
# given that they have embarked or boarded from a particular port

sibSpQ = titanic[titanic.Embarked == 'Q'][titanic.SibSp== 1].shape[0]
sibSpC = titanic[titanic.Embarked == 'C'][titanic.SibSp == 1].shape[0]
sibSpS = titanic[titanic.Embarked == 'S'][titanic.SibSp == 1].shape[0]
print(sibSpQ)
print(sibSpC)
print(sibSpS)
```

```python
sibSpQ = titanic[titanic.Embarked == 'Q'][titanic.SibSp == 0].shape[0]
sibSpC = titanic[titanic.Embarked == 'C'][titanic.SibSp == 0].shape[0]
sibSpS = titanic[titanic.Embarked == 'S'][titanic.SibSp == 0].shape[0]
print(sibSpQ)
print(sibSpC)
print(sibSpS)
```

```python
titanic.dropna(inplace=True)
titanic.head()
```

```python
titanic.isnull().sum()
```

```python
#Renaming "age" and "gender" columns
titanic.rename(columns={'age':'Age'}, inplace=True)
titanic.head()
```

```python
titanic.rename(columns={'Gender':'Sex'}, inplace=True)
titanic.head()
```

```python
def getEmb(str):
    if str=="S":
        return 1
    elif str=='Q':
        return 2
    else:
        return 3
titanic["Embark"]=titanic["Embarked"].apply(getEmb)
titanic.head()
```

```python
del titanic['Embarked']
titanic.rename(columns={'Embark':'Embarked'}, inplace=True)
titanic.head()
```

```python
#Drawing a pie chart for number of males and females aboard
import matplotlib.pyplot as plt
from matplotlib import style

males = (titanic['Sex'] == 1).sum() 
#Summing up all the values of column gender with a 
#condition for male and similary for females
females = (titanic['Sex'] == 2).sum()
print(males)
print(females)
p = [males, females]
plt.pie(p,    #giving array
       labels = ['Male', 'Female'], #Correspndingly giving labels
       colors = ['green', 'yellow'],   # Corresponding colors
       explode = (0.15, 0),    #How much the gap should me there between the pies
       startangle = 0)  #what start angle should be given
plt.axis('equal') 
plt.show()
```

```python
# More Precise Pie Chart
MaleS=titanic[titanic.Sex==1][titanic.SibSp==1].shape[0]
print(MaleS)
MaleN=titanic[titanic.Sex==1][titanic.SibSp==0].shape[0]
print(MaleN)
FemaleS=titanic[titanic.Sex==2][titanic.SibSp==1].shape[0]
print(FemaleS)
FemaleN=titanic[titanic.Sex==2][titanic.SibSp==0].shape[0]
print(FemaleN)
```

```python
chart=[MaleS,MaleN,FemaleS,FemaleN]
colors=['lightskyblue','yellowgreen','Yellow','Orange']
labels=["Survived Male","Not Survived Male","Survived Female","Not Survived Female"]
explode=[0,0.05,0,0.1]
plt.pie(chart,labels=labels,colors=colors,explode=explode,startangle=100,counterclock=False,autopct="%.2f%%")
plt.axis("equal")
plt.show()
```
